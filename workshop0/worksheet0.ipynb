**3. Sum of All Elements in a Nested List:**

```python

def sum_nested(lst):
  sum = 0
  for x in lst:
    if isinstance(x,list):
      for y in x:
        if isinstance(y,list):
          for z in y:
            sum = sum + z
        else:
          sum = sum + y
    else:
      sum = sum + x
  print(sum)
lst = [[1, 2], [3, [4, 5]], 6]
sum_nested(lst)
```

```python

def remove(lst,n):
  b=[]
  for x in lst:
    if isinstance(x, list):
      b.append(remove(x, n))
    else:
      if x != n:
        b.append(x)
  return(b)
a = [[1, 2], [3, [4, 5]], 6]
print(remove(a,2))
```

**4. Matrix Multiplication**

```python
import time

n = 1000

# Create two 1000x1000 matrices using lists
A = [[i for i in range(n)] for _ in range(n)]
B = [[j for j in range(n)] for _ in range(n)]

# Matrix multiplication with lists
start = time.time()
result_list = [[sum(A[i][k] * B[k][j] for k in range(n)) for j in range(n)] for i in range(n)]
end = time.time()

print("Time taken with Python lists:", end - start, "seconds")
```
